```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='white', palette='Set2')
```

```python
df = pd.read_csv('../data/fandango_scrape.csv')
sns.scatterplot(data=df, x = 'RATING', y = 'VOTES')
```

```python
df.corr(numeric_only=True)
```

```python
df = pd.read_csv('../data/fandango_scrape.csv')
df = df[df['VOTES'] > 0]
sns.kdeplot(df["STARS"], label="STARS", clip=(0,5), bw_adjust=1.5, fill=True)
sns.kdeplot(df["RATING"], label="RATING", clip=(0,5), bw_adjust=1.5, fill=True)
plt.legend()
```

```python
df = pd.read_csv('../data/fandango_scrape.csv')
df["YEAR"] = df["FILM"].str.extract(r"\((\d{4})\)").astype(int)
sns.countplot(data=df, x='YEAR')
```

```python
df = pd.read_csv('../data/fandango_scrape.csv')
df = df[df['VOTES'] > 0]
df['STARS_DIFF'] = (df['STARS'] - df['RATING']).round(1)
sns.countplot(data=df, x = 'STARS_DIFF', palette='Set2')
```
